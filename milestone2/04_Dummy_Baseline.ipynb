# Dummy Baseline Model

This notebook creates a baseline model by extracting the first 1000 characters of the abstracts.

```python
import pandas as pd
```

```python
from google.colab import drive

drive.mount("/content/drive")
```

# Load Validation Data

```python
INPUT_FILE_DIR = "/content/drive/MyDrive/BioLaySumm2024_main/extracted/"
ELIFE_VAL_FILE_NAME = "eLife_val_extracted.csv"
PLOS_VAL_FILE_NAME = "PLOS_val_extracted.csv"

val_eLife_df = pd.read_csv(INPUT_FILE_DIR + ELIFE_VAL_FILE_NAME)

val_plos_df = pd.read_csv(INPUT_FILE_DIR + PLOS_VAL_FILE_NAME)
```

```python
val_eLife_df['abstract'].apply(lambda x: x[:1000])
```

# Output Validation data

```python
OUTPUT_FILE_DIR = "/content/drive/MyDrive/BioLaySumm2024-evaluation_scripts/baseline_evaluation/"

val_eLife_df['abstract'].apply(lambda x: x[:1000]).to_csv(OUTPUT_FILE_DIR+'elife.txt', index=False, header=False, sep='\n')

val_plos_df['abstract'].apply(lambda x: x[:1000]).to_csv(OUTPUT_FILE_DIR+'plos.txt', index=False, header=False, sep='\n')
```

# Load test data

```python
ELIFE_TEST_FILE_NAME = "eLife_test_extracted.csv"
PLOS_TEST_FILE_NAME = "PLOS_test_extracted.csv"


test_eLife_df = pd.read_csv(INPUT_FILE_DIR + ELIFE_TEST_FILE_NAME)

test_plos_df = pd.read_csv(INPUT_FILE_DIR + PLOS_TEST_FILE_NAME)
```

# Output Test Data

```python
OUTPUT_FILE_DIR = "/content/drive/MyDrive/BioLaySumm2024_main/test_submissions/dummy_baseline_submission/"

test_eLife_df['abstract'].apply(lambda x: x[:1000]).to_csv(OUTPUT_FILE_DIR+'elife.txt', index=False, header=False, sep='\n')

test_plos_df['abstract'].apply(lambda x: x[:1000]).to_csv(OUTPUT_FILE_DIR+'plos.txt', index=False, header=False, sep='\n')
```
